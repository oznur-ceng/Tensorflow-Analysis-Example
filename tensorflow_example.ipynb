```python
import pandas as pd
import numpy as np
```

```python
dataFrame=pd.read_excel("maliciousornot.xlsx")
```

```python
dataFrame
```

```python
dataFrame.info()
```

```python
dataFrame.describe()
```

```python
y=dataFrame["Type"].values
x=dataFrame.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model=Sequential()
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
modelKaybı=pd.DataFrame(model.history.history)
```

```python
modelKaybı.plot()
```

```python
model=Sequential()
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKaybı=pd.DataFrame(model.history.history)
```

```python
modelKaybı.plot()
```

```python
model=Sequential()
model.add(Dense(units=30,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(units=1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
kayıpDf=pd.DataFrame(model.history.history)
```

```python
kayıpDf.plot()
```

```python
tahminVeri=model.predict_classes(x_test)
```

```python
tahminVeri
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,tahminVeri))
```

```python
print(confusion_matrix(y_test,tahminVeri))
```
